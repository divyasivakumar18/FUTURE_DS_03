```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# 1. Upload the file
print("Please upload 'feedback3.csv'")
uploaded = files.upload()

# 2. Load the dataset
# Based on the file structure, it contains pairs of ratings and text comments
df = pd.read_csv('feedback3.csv')

# 3. Clean column names (removing leading/trailing spaces)
df.columns = df.columns.str.strip()

# 4. Display basic information
print("\n--- Dataset Info ---")
print(df.info())

print("\n--- First 5 Rows ---")
display(df.head())

# 5. Summary of Numerical Ratings
# The ratings are in 'teaching', 'coursecontent', 'examination', 'labwork', 'library_facilities', 'extracurricular'
rating_cols = ['teaching', 'coursecontent', 'examination', 'labwork', 'library_facilities', 'extracurricular']

print("\n--- Average Ratings ---")
# Filter the DataFrame to include only numeric columns that are in rating_cols
# This handles duplicate column names by ensuring only numeric dtypes are kept for calculation.
numeric_rating_df = df[rating_cols].select_dtypes(include=['number'])
print(numeric_rating_df.mean())

# 6. Basic Visualization
plt.figure(figsize=(12, 6))
# Use the numeric_rating_df for melting as well to ensure consistency
df_melted = numeric_rating_df.melt(var_name='Category', value_name='Rating')
# Updated sns.barplot call to address FutureWarning
sns.barplot(x='Category', y='Rating', hue='Category', data=df_melted, estimator='mean', palette='viridis', legend=False)
plt.title('Average Ratings per Category')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.show()
```
